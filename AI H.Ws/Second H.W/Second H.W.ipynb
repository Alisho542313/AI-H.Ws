```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get("https://www.bbc.com/")
```

```python
res
```

```python
res.text
```

```python
soup = BeautifulSoup(res.text,'html.parser')
```

```python
soup
```

```python
title = soup.select('.sc-8ea7699c-3')
```

```python
title
```

```python
news=[]
for i,v in enumerate(title):
    print(i+1,'-',v.text)
    news.append(v.text)
```

```python
news
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df['title'] = news
```

```python
df
```

```python
df.to_csv('News2.csv',index=False)
```
