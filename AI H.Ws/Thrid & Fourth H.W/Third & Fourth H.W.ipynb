```python
import pandas as pd
```

```python
data = pd.read_csv("train.tsv",delimiter="\t")
```

```python
data = data.dropna()
```

```python
data.shape
```

```python
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
```

```python
print(data['selected_text'])
```

```python
data['selected_text'] = data['selected_text'].apply(lambda x:' '.join([porter.stem(i) for i in x.split(' ')]))
```

```python
print(data['selected_text'])
```

```python
import string
```

```python
punc = string.punctuation
punc
```

```python
table = str.maketrans('','',punc)
```

```python
data['selected_text'] = data['selected_text'].apply(lambda x: x.translate(table))
```

```python
print(data['selected_text'])
```

```python
x = data['selected_text']
y = data['sentiment']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
nv_classifier = Pipeline([("tf-idf",TfidfVectorizer()),("nvbayses",MultinomialNB())])
svc_classifier = Pipeline([('Tf-idf',TfidfVectorizer()),('svc',LinearSVC())])
```

# Naive-Bayes


```python
nv_classifier.fit(x_train,y_train)
```

```python
predictions = nv_classifier.predict(x_test)
```

```python
accuracy_score(y_test,predictions)
```

```python
nv_classifier.predict(['this game is cool'])
```

```python
nv_classifier.fit(x_train, y_train)
```

```python
nv_predictions = nv_classifier.predict(x_test)
```

```python
nv_accuracy = accuracy_score(y_test, nv_predictions)
```
